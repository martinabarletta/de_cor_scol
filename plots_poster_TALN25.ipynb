```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# Données
pays = ['français', 'italien', 'espagnol']
cm2 = [336, 17, 585]
cm1 = [369, 0, 457]
ce2 = [369, 96, 394]
ce1 = [373, 672, 382]
cp = [373, 702, 382]

# Palette pastel
colors = sns.color_palette("pastel")
color_list = colors[:5]

# Inverser l'ordre pour empiler du bas vers le haut
levels = [cm2, cm1, ce2, ce1, cp]
labels = ['5 - CM2', '4 - CM1', '3 - CE2', '2 - CE1', '1 - CP']
legend_labels = list(reversed(labels))
legend_colors = list(reversed(color_list))

# Créer la figure
fig, ax = plt.subplots(figsize=(10, 8))
fig.patch.set_facecolor('white')  # Fond blanc

# Empilement des barres
bottom_vals = np.zeros(len(pays))
bars = []
for i, values in enumerate(levels):
    bar = ax.bar(pays, values, bottom=bottom_vals, color=color_list[i], label=labels[i], width=0.5)
    bars.append(bar)
    bottom_vals += values

# Afficher les totaux au-dessus des colonnes
for i, total in enumerate(bottom_vals):
    ax.text(i, total + 100, f'{int(total)}', ha='center', va='bottom', fontsize=26)

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height >= 20:
            # Valore normale dentro la barra
            x = bar.get_x() + bar.get_width() / 2.
            y = bar.get_y() + height / 2.
            ax.text(x, y, f'{int(height)}', ha='center', va='center', fontsize=22)
        elif height > 0:
            # Valore piccolo: accanto a sinistra, dentro il plot
            x = bar.get_x() - 0.1  # leggermente a sinistra del bordo sinistro della barra
            y = bar.get_y() + height / 2.
            ax.plot([bar.get_x(), x + 0.03], [y, y], color='black', linewidth=1)  # stanghetta
            ax.text(x, y, f'{int(height)}', ha='right', va='center', fontsize=22)

for b in bars:
    add_labels(b)

# Paramètres de l’axe
ax.set_title('Textes par langue et par niveau scolaire', fontsize=30)
ax.set_ylabel('Nb textes', fontsize=24)
ax.set_ylim(0, 2800)
ax.set_yticks(np.arange(0, 2500, 500))
ax.tick_params(axis='both', labelsize=20)

ax.grid(False)
ax.grid(axis='y', linestyle=':', alpha=0.7)
# Légende à l’intérieur, coin supérieur gauche
legend_patches = [Patch(color=c, label=l) for c, l in zip(legend_colors, legend_labels)]
legend = ax.legend(handles=legend_patches, loc='upper left', frameon=True, prop={'size': 15})
legend.get_frame().set_edgecolor('black')
legend.get_frame().set_linewidth(1)

# Ajustement du layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Caricamento dati
df_fr = pd.read_csv("./annotations_francais/corpus_francais_v3.csv")
df_it = pd.read_csv("./annotations_italien/corpus_italien.csv")

df_fr["langue"] = "français"
df_it["langue"] = "italien"
df_fr["niveau"] = df_fr["texte"].str.extract(r"(CE1|CE2)")
df_it["niveau"] = df_it["texte"].str.extract(r"(CE1|CE2)")

df_fr_filtered = df_fr[["nbTokNoPunct", "langue", "niveau"]].dropna()
df_it_filtered = df_it[["nbTokNoPunct", "langue", "niveau"]].dropna()
df_combined = pd.concat([df_fr_filtered, df_it_filtered])
df_combined["niveau"] = pd.Categorical(df_combined["niveau"], categories=["CE1", "CE2"])

# Palette
full_palette = sns.color_palette("pastel")
selected_colors = [full_palette[6], full_palette[4]]

# Plot
plt.figure(figsize=(8, 6))
ax = sns.boxplot(data=df_combined, x="niveau", y="nbTokNoPunct", hue="langue", palette=selected_colors)
fig.patch.set_facecolor('white')  # Fond blanc

label=14

# Calcolo mediana per ogni gruppo
grouped = df_combined.groupby(['niveau', 'langue'])['nbTokNoPunct'].median().reset_index()

# Posizioni dei box stimati manualmente
# seaborn assegna automaticamente le posizioni dei gruppi come: 0, 1 per 'CE1' e 'CE2'
# all’interno di ciascun gruppo, i livelli hue (français, italien) sono distanziati di +/- offset
x_offsets = {'français': -0.2, 'italien': 0.2}
x_base = {'CE1': 0, 'CE2': 1}

for _, row in grouped.iterrows():
    niveau = row['niveau']
    langue = row['langue']
    median = row['nbTokNoPunct']

    x = x_base[niveau] + x_offsets[langue]
    ax.text(x, median + 10, f'{int(median)}',
            ha='center', va='bottom', fontsize=label, color='black',
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2'))
# Calcola min e max per ogni box
grouped_extremes = df_combined.groupby(['niveau', 'langue'])['nbTokNoPunct'].agg(['min', 'max']).reset_index()

for _, row in grouped_extremes.iterrows():
    niveau = row['niveau']
    langue = row['langue']
    val_min = row['min']
    val_max = row['max']

    x = x_base[niveau] + x_offsets[langue]

    # Etichetta sotto il minimo
    ax.text(x, val_min - 10, f'{int(val_min)}',
            ha='center', va='top', fontsize=label, color='black',
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2'))

    # Etichetta sopra il massimo
    ax.text(x, val_max + 10, f'{int(val_max)}',
            ha='center', va='bottom', fontsize=label, color='black',
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2'))

# Titoli e legenda
plt.title("Comparaison de la longueur des textes \npar langue et par niveau", fontsize=16, pad=30)
plt.ylabel("Nombre de tokens par texte", fontsize=14)
plt.xlabel("Niveau", fontsize=14)
ax.tick_params(axis='both', labelsize=14)

leg = ax.legend(title="Langue", prop={'size': 12})
leg.get_title().set_fontsize(12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Funzione per costruire la matrice posizione × tipo ---
def build_position_type_matrix(dfs_dict, sheet_groups, max_position=5):
    """
    Costruisce una matrice (DataFrame) con:
    - righe: posizione nella catena (tag_occurrences)
    - colonne: tipo di menzione (es. 'Pronom', 'SN défini', ...)
    - valori: conteggio delle occorrenze
    """
    records = []

    for sheet_name, df in dfs_dict.items():
        if 'tag_occurrences' in df.columns:
            df = df.dropna(subset=['tag_occurrences'])  # rimuove righe senza posizione
            df['tag_occurrences'] = pd.to_numeric(df['tag_occurrences'], errors='coerce')
            df = df[df['tag_occurrences'] <= max_position]  # limita fino a max_position

            group_type = sheet_groups.get(sheet_name, 'Autre')
            counts = df['tag_occurrences'].value_counts()
            for pos, count in counts.items():
                records.append({'Position': int(pos), 'Type': group_type, 'Count': count})

    matrix_df = pd.DataFrame(records)
    pivot = matrix_df.pivot_table(index='Position', columns='Type', values='Count', aggfunc='sum', fill_value=0)

    # Ordina le colonne secondo l'ordine preferito
    categories = ["Pronom", "SN défini", "SN indéfini", "SN possessif", "SN démonstratif",
                  "SN sans déterminant", "Nom Propre", "Anaphore Zéro", "Autre"]
    pivot = pivot.reindex(columns=categories, fill_value=0)

    return pivot.sort_index()

# --- Funzione per plottare la heatmap ---
def plot_heatmap(matrix_df, language='Français', max_position=5):
    plt.figure(figsize=(12, 8))
    sns.heatmap(matrix_df, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.5, cbar_kws={'label': 'Occurrences'})
    plt.title(f"Distribution des types de mentions par position ({language})", fontsize=16)
    plt.ylabel("Position dans la chaîne", fontsize=14)
    plt.xlabel("Type de mention", fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f'./heatmap_typologie_{language.lower()}.png', dpi=1200, bbox_inches='tight')
    plt.show()

# --- Creazione matrici e plot ---
matrix_fr = build_position_type_matrix(filtered_fr, sheet_groups, max_position=5)
matrix_it = build_position_type_matrix(filtered_ita, sheet_groups, max_position=5)

plot_heatmap(matrix_fr, language='Français')
plot_heatmap(matrix_it, language='Italien')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def build_position_type_matrix_transposed(dfs_dict, sheet_groups, max_position=10):
    """
    Costruisce una matrice con:
    - righe: tipo di menzione (es. 'Pronom', 'SN défini', ...)
    - colonne: posizione nella catena (1, 2, ..., max_position)
    - valori: conteggio delle occorrenze
    """
    records = []

    for sheet_name, df in dfs_dict.items():
        if 'tag_occurrences' in df.columns:
            df = df.dropna(subset=['tag_occurrences'])
            df['tag_occurrences'] = pd.to_numeric(df['tag_occurrences'], errors='coerce')
            df = df[(df['tag_occurrences'] >= 1) & (df['tag_occurrences'] <= max_position)]

            group_type = sheet_groups.get(sheet_name, 'Autre')
            counts = df['tag_occurrences'].value_counts()
            for pos, count in counts.items():
                records.append({'Type': group_type, 'Position': int(pos), 'Count': count})

    matrix_df = pd.DataFrame(records)
    pivot = matrix_df.pivot_table(index='Type', columns='Position', values='Count', aggfunc='sum', fill_value=0)

    # Ordina le righe secondo l'ordine delle categorie
    categories = ["Pronom", "SN défini", "SN indéfini", "SN possessif", "SN démonstratif",
                  "SN sans déterminant", "Nom Propre", "Anaphore Zéro", "Autre"]
    pivot = pivot.reindex(categories, fill_value=0)

    # Ordina le colonne (posizioni da 1 a max_position)
    pivot = pivot.reindex(columns=range(1, max_position+1), fill_value=0)

    return pivot
    
def plot_heatmaps_side_by_side(matrix_fr, matrix_it, max_position=10,
                               figsize=(20,10),
                               title_fontsize=18,
                               label_fontsize=14,
                               tick_fontsize=12):
    fig, axes = plt.subplots(1, 2, figsize=figsize, sharey=True)
    # Ordina le righe secondo l'ordine delle categorie
    categories = ["Pronom", "SN défini", "SN indéfini", "SN possessif", "SN démonstratif",
                  "SN sans déterminant", "Nom Propre", "Anaphore Zéro", "Autre"]

    cmap = "YlGnBu"

    sns.heatmap(matrix_fr, annot=True, fmt="d", cmap=cmap, linewidths=0.3,
                cbar=False, square=True, ax=axes[0])

    axes[0].set_title("Distribution des types de mentions par position (Français)", fontsize=title_fontsize)
    axes[0].set_xlabel("Position dans la chaîne", fontsize=label_fontsize)
    axes[0].set_ylabel("Type de mention", fontsize=label_fontsize)
    axes[0].tick_params(axis='x', labelsize=tick_fontsize)
    axes[0].tick_params(axis='y', labelsize=tick_fontsize)
    axes[0].set_xticklabels([str(x) for x in range(1, max_position+1)])

    # Forza ticks y e labels y
    axes[0].set_yticks(range(len(categories)))
    axes[0].set_yticklabels(categories, rotation=0, fontsize=tick_fontsize)


    # Heatmap per il plot italiano, senza colorbar (la lasciamo solo a sinistra)
    sns.heatmap(matrix_it, annot=True, fmt="d", cmap=cmap, linewidths=0.3,
                cbar=True, cbar_kws={'label': 'Occurrences'}, square=True, ax=axes[1])
    axes[1].set_title("Distribution des types de mentions par position (Italien)", fontsize=title_fontsize)
    axes[1].set_xlabel("Position dans la chaîne", fontsize=label_fontsize)
    axes[1].set_ylabel("")  # niente label y qui
    axes[1].tick_params(axis='x', labelsize=tick_fontsize)
    axes[1].tick_params(axis='y', labelsize=tick_fontsize)  # rimuove ticks asse y nel plot di destra
    axes[1].set_yticklabels([])  # nasconde etichette y nel plot di destra
    axes[1].set_xticklabels([str(x) for x in range(1, max_position+1)])

    plt.tight_layout()
    plt.savefig('./heatmap_typologie_fr_it_side_by_side.png', dpi=1200, bbox_inches='tight')
    plt.show()


# --- Creazione matrici ---
max_position = 10
matrix_fr = build_position_type_matrix_transposed(filtered_fr, sheet_groups, max_position=max_position)
matrix_it = build_position_type_matrix_transposed(filtered_ita, sheet_groups, max_position=max_position)

# --- Plot affiancati ---
plot_heatmaps_side_by_side(matrix_fr, matrix_it, max_position=max_position)
```
