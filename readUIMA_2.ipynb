Ce code prends en entrée un fichier xmi et son fichier TypeSystem en xml et permet de lire les annotations contenues

```python
import pandas as pd
from cassis import *
import json
```

```python
#ouvrir le fichier TypeSystem qui est en xml
#remplacer "myfolder" avec l'emplacement du fichier
with open('./barlettm/TypeSystem.xml', 'rb') as f:
    typesystem = load_typesystem(f)
```

```python
#ouvrir le fichier xmi
def load_cas_xmi(file_path):
    """
    Load a CAS XMI file without a predefined type system and with leniency turned on.
    """
    with open(file_path, 'rb') as f:
        cas = load_cas_from_xmi(f, typesystem=typesystem)  # , lenient=True Leniency helps bypass MetaData errors
    return cas
#utilisation :
file_path = './barlettm/barlettm.xmi'

# Load CAS with leniency
cas = load_cas_xmi(file_path)
```

```python
# Assuming `cas` is the loaded CAS object
json_output = cas.to_json(
    path=None,            # Return as a string, not a file
    pretty_print=True,    # Pretty-print the JSON (optional)
    ensure_ascii=False,   # Don't escape non-ASCII characters
    #type_system_mode="MINIMAL"  # Exclude type system information, or use "FUL" or "MINIMAL"
)
```

```python
# see all annotations availables in xmi file
vista = cas._current_view.get_all_annotations()
print(vista)
```

```python
#celles la sont les annotations dont on a besoin pour votre projet, les noms des layers disponibles 
#dans les annotations sont contenus dans le fichier TypeSystem.xml
```

```python
#access mentions custom webanno annotations
for sentence in cas.select('de.tudarmstadt.ukp.dkpro.core.api.segmentation.type.Sentence'):
    for token in cas.select_covered('webanno.custom.Mentions', sentence):
        #position de la mention annotée en caractères, texte couvert par l'annotation
        print(token, token.get_covered_text())
```

```python
#access coref custom webanno annotations
for sentence in cas.select('de.tudarmstadt.ukp.dkpro.core.api.segmentation.type.Sentence'):
    for token in cas.select_covered('webanno.custom.Coref', sentence):
        print(token, token.get_covered_text())
```
