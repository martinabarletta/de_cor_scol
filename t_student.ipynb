```python
# -*- coding: utf-8 -*-
"""
Created on Mon May 19 22:49:58 2025

@author: matil

1. comparaison entre dénsité réferentielle moyenne de chaque texte (contenue dans density_fr)
"""
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

res_fr = "./annotations_francais/density_fr.csv"
res_it = "./annotations_italien/density_ita.csv"

# Carica il file CSV
df1 = pd.read_csv(res_fr)
df2 = pd.read_csv(res_it)
# Visualizza le prime righe per verificare i dati
print(df1.head())

#t di student tra densità per lingua (fr vs it)
gruppo1 = df1['density']
gruppo2 = df2['density']

# Esegui il test t
statistica, p_value = ttest_ind(gruppo1, gruppo2, equal_var=False)  # equal_var=False per il test di Welch

print(f'Statistica t: {statistica}')
print(f'P-value: {p_value}')
#####tra densità in francese e in italiano :
    ## Statistica t: -0.4524183694398316
    ##P-value: 0.6513043013512673 - différence significative ?
# 📊 Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([gruppo1, gruppo2], labels=['français', 'italien'])
plt.title(f'Comparaison densité référentielle entre français et italien\nTest t: p = {p_value:.3g}')
plt.ylabel('Densité référentielle')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
##entre niveaux pour chaque langue ex CE1 et CE2 français
#t di student tra densità per lingua
gruppo1 = df1[df1['group'] == 'CE1']['density']
gruppo2 = df1[df1['group'] == 'CE2']['density']

# Esegui il test t
statistica, p_value = ttest_ind(gruppo1, gruppo2, equal_var=True)  # equal_var=False per il test di Welch
print(f'Statistica t: {statistica}')
print(f'P-value: {p_value}')

#####tra densità in francese su CE1 e CE2
# 📊 Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([gruppo1, gruppo2], labels=['CE1', 'CE2'])
plt.title(f'Comparaison densité référentielle entre CE1 et CE2 français\nTest t: p = {p_value:.3g}')
plt.ylabel('Densité référentielle')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
##entre niveaux pour chaque langue ex CE1 et CE2 italien
#t di student tra densità per lingua
gruppo1 = df2[df2['group'] == 'CE1']['density']
gruppo2 = df2[df2['group'] == 'CE2']['density']

# Esegui il test t
statistica, p_value = ttest_ind(gruppo1, gruppo2, equal_var=False)  # equal_var=False per il test di Welch
print(f'Statistica t: {statistica}')
print(f'P-value: {p_value}')

#####tra densità in francese e in italiano :
    ## Statistica t: -0.4524183694398316
    ##P-value: 0.6513043013512673 - différence significative ?
# 📊 Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([gruppo1, gruppo2], labels=['CE1', 'CE2'])
plt.title(f'Comparaison densité référentielle entre CE1 et CE2 italien\nTest t: p = {p_value:.3g}')
plt.ylabel('Densité référentielle')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
